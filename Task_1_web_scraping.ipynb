# INFINIX MOBILE PHONE DATA SCRAPES FROM DARAZ
For Web scraping I couldnt able to Scrape DARAZ website from the url so for data collection i have downloaded the DARAZ web page and from the downloaded web page i completed the remaining scraping task.


```python
from bs4 import BeautifulSoup
import pandas as pd

# Open and read the HTML file
with open("Buy Infinix Smartphones at Best Price in Pakistan - (April, 2023) - Daraz.pk.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all product containers with the specified class
product_containers = soup.find_all('div', class_='info--ifj7U')

# Create an empty list to store the data
data = []

# Loop through each product container to extract the title and price
for product in product_containers:
    # Extract product title
    title = product.find('div', class_='title--wFj93').text.strip()

    # Extract product price
    price = product.find('div', class_='price--NVB62').text.strip()

    # Append the data to the list
    data.append([title, price])

# Create a Pandas DataFrame from the collected data
daraz_df = pd.DataFrame(data, columns=["Product Title", "Daraz Product Price"])

# Display the DataFrame
daraz_df.head(10)
```

```python
# Remove "Rs." from the 'Product Price' column
daraz_df['Daraz Product Price'] = daraz_df['Daraz Product Price'].str.replace('Rs.', '').str.strip()
# Extract the first three words from the 'Product Title' column and assign it to a new column 'Model'
daraz_df['Model'] = daraz_df['Product Title'].str.split().str[:3].str.join(' ')
daraz_df.head(10)
```

# INFINIX MOBILE PHONE DATA SCRAPES FROM OLX
For Web scraping I couldnt able to Scrape olx website from the Url so for data collection i have downloaded the olx web page and from the downloaded web page i completed the remaining scraping task.

```python
from bs4 import BeautifulSoup

# Open and read the HTML file
with open("Infinix... - Mobile Phones for sale in Pakistan _ OLX Pakistan.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all product containers with the specified class for product title
product_containers = soup.find_all('div', class_='a5112ca8 _5fdf4379')

# Create an empty list to store the data
data = []

# Loop through each product container to extract the title and price
for product in product_containers:
    # Extract product title
    title = product.text.strip()

    # Extract product price
    price = product.find_next('div', {'aria-label': 'Price'}).text.strip()

    # Append the data to the list
    data.append([title, price])

# Create a Pandas DataFrame from the collected data
import pandas as pd

olx_df = pd.DataFrame(data, columns=["Product Title", "Olx Product Price"])

# Display the DataFrame
olx_df.head(10)
```

```python
# Extract the first three words from the 'Product Title' column and assign it to a new column 'Model'
olx_df['Model'] = olx_df['Product Title'].str.split().str[:3].str.join(' ')
# Filter rows where the 'Model' column starts with "Infinix"
olx_df = olx_df[olx_df['Model'].str.startswith('Infinix')]
# Remove "Rs." from the 'Product Price' column
olx_df['Olx Product Price'] = olx_df['Olx Product Price'].str.replace('Rs.', '').str.strip()
# Display the updated DataFrame
olx_df.head(10)
```

# Explanation
That isn't perfect comparison between both product as in Olx we have difference types of sale occurs for example we can get used mobilephone from there too. and they have posted for Loan Prices too so that's why loan prices are also included in comparison of both.


```python
df_olx_subset = olx_df[['Model', 'Olx Product Price']]
df_daraz_subset = daraz_df[['Model', 'Daraz Product Price']]

# Merging DataFrames on the 'Model' column
merged_df = pd.merge(df_olx_subset, df_daraz_subset, on='Model')

# Displaying the merged DataFrame
print(merged_df)
```
